```python
# Libraries
import pandas as pd
```

```python
# loading dataset and reading as a dataframe
df=pd.read_csv('/content/drive/MyDrive/Salary_dataset.csv')
df.head(5)
```

```python
# checking Null values
df.isnull().sum()
```

```python
# i. fillna
df.fillna(0)
```

```python
# ii. dropna
df.dropna()
```

```python
# iii. replace
df.replace(3.1,3.0 ,inplace=True)
df.replace(3.2,3.0 ,inplace=True)
df.head(20)
```

```python
# iv. groupby
grouped=df.groupby('YearsExperience')
result = grouped['Salary'].sum()

print(result)
```

```python
# v. concat
df2=pd.read_csv('/content/drive/MyDrive/USA_Housing.csv')

# Concantate df1 and df2
result = pd.concat([df, df2], axis=0)
print(result)
```

```python
df3=pd.read_csv('/content/drive/MyDrive/Salary_dataset.csv')
```

```python
# vi. merge (use it with inner, outer, left, right)
# Inner join
inner_join = pd.merge(df, df3, on='Salary', how='inner')
print("nner Join:")
print(inner_join)

# Outer join
outer_join = pd.merge(df, df3, on='Salary', how='outer')
print("\nOuter Join:")
print(outer_join)

# Left join
left_join = pd.merge(df, df3, on='Salary', how='left')
print("\nLeft Join:")
print(left_join)

# Right join
right_join = pd.merge(df, df3, on='Salary', how='right')
print("\nRight Join:")
print(right_join)
```

```python
# vii. pivot and pivot_table
# Using pivot
pivot = df.pivot(index='Salary', columns='YearsExperience', values='3.0')
print("Pivot Table:")
print(pivot)

# Using pivot_table
pivot_table = df.pivot_table(index='index', columns='YearsExperience', values='3.0')
print("\nPivot Table:")
print(pivot_table)
```

```python
# viii. melt
melted_df = df.melt(id_vars=['Salary'], value_vars=['YearsExperience', 'Salary'], 
                    var_name='Variable', value_name='Value')

print(melted_df)
```

```python
# ix. stack and unstack
stacked_df = df.set_index(['Salary']).stack()

print(stacked_df)
```

```python
# x. crosstab
table = pd.crosstab(index=df['Salary'], columns=df['Salary'], values=df['YearsExperience'], aggfunc='sum')

print(table)
```
